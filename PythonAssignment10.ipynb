```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from CSV file
data = pd.read_csv('BigMartSalesData.csv')

# Convert the 'InvoiceDate' column to datetime
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])

# Extract year and month from 'InvoiceDate'
data['Year'] = data['InvoiceDate'].dt.year
data['Month'] = data['InvoiceDate'].dt.month

# Group data by year and month to calculate total sales per month
monthly_sales = data[data['Year'] == 2011].groupby('Month')['Amount'].sum()

# Plot Total Sales Per Month for Year 2011
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title('Total Sales Per Month for Year 2011')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()

# Plot Total Sales Per Month for Year 2011 as Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(monthly_sales.index, monthly_sales.values)
plt.title('Total Sales Per Month for Year 2011 (Bar Chart)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()

# Plot Pie Chart for Year 2011 Country Wise
country_sales = data[data['Year'] == 2011].groupby('Country')['Amount'].sum()
plt.figure(figsize=(8, 8))
plt.pie(country_sales, labels=country_sales.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Sales Distribution by Country for Year 2011')
plt.show()

# Plot Scatter Plot for the invoice amounts
plt.figure(figsize=(10, 6))
plt.scatter(data['InvoiceNo'], data['Amount'], alpha=0.5)
plt.title('Scatter Plot of Invoice Amounts')
plt.xlabel('Invoice Number')
plt.ylabel('Amount')
plt.show()
```

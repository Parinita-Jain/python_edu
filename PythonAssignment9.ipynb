```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Hurricanes.csv')

# Plot the bar graph
plt.figure(figsize=(10, 6))
plt.bar(data['Year'], data['Hurricanes'])
plt.xlabel('Year')
plt.ylabel('Number of Hurricanes')
plt.title('Number of Hurricanes in the United States (Atlantic Coast)')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('CityTemps.csv')

# Filter data for San Francisco and Moscow
sf_temps = data['San Francisco']
moscow_temps = data['Moscow']

# Plot histograms
plt.figure(figsize=(10, 6))
plt.hist(sf_temps, bins=20, alpha=0.5, label='San Francisco', color='blue')
plt.hist(moscow_temps, bins=20, alpha=0.5, label='Moscow', color='orange')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Temperature Histograms for San Francisco and Moscow')
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Cars2015.csv')

# Group data by manufacturer and count models
manufacturer_counts = data['Make'].value_counts()

# Plot pie chart
plt.figure(figsize=(16, 16))
plt.pie(manufacturer_counts, labels=manufacturer_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Number of Models Released by Manufacturer')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Find the manufacturer with the largest releases
largest_manufacturer = manufacturer_counts.idxmax()
print(f"The manufacturer with the largest releases is: {largest_manufacturer}")

# Show the plot
plt.show()
```

```python
import pandas as pd

# Read CSV data into a DataFrame
data = pd.read_csv('sales_data.csv')

# Display the data
data
```

```python
# Describe the data on unit_price
unit_price_description = data['unit price'].describe()
print(unit_price_description)
```

```python
# Create a new DataFrame with selected columns and group by name
grouped_data = data[['name', 'unit price', 'date']].groupby('name').sum()

# Display the grouped data
print(grouped_data)
```

```python
import matplotlib.pyplot as plt

# Calculate total sales by each customer
total_sales = grouped_data['unit price']

# Plot the graph
plt.figure(figsize=(10, 6))
total_sales.plot(kind='bar')
plt.xlabel('Customer Name')
plt.ylabel('Total Sales')
plt.title('Total Sales by Customer')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the DataFrame
data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'female': [0, 1, 1, 0, 1],
    'age': [42, 52, 36, 24, 73],
    'preTestScore': [4, 24, 31, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70]
}
df = pd.DataFrame(data)

# Create a color map based on gender (0: male, 1: female)
colors = ['blue' if gender == 0 else 'pink' for gender in df['female']]

# Scatter plot with size determined by 'age' and color determined by 'sex'
plt.figure()
plt.scatter(df['preTestScore'], df['postTestScore'], s=300, c=colors, alpha=0.7)

# Add labels and title
plt.title("Scatterplot of preTestScore vs postTestScore")
plt.xlabel("preTestScore")
plt.ylabel("postTestScore")

# Show the plot
plt.show()
```
